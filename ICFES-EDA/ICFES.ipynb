Podemos ver que las variables de las discapacidades son casi nulas en su totalidad. Ya que las variables de discapacidad especifican las discapacidades, vamos a tomar los valores nulos como que no posee discapacidades.

```python
columnas_discapacidades = ['ESTU_LIMITA_AUTISMO', 'ESTU_LIMITA_SORDO', 'ESTU_LIMITA_CONDICIONESPECIAL', 'ESTU_LIMITA_INVIDENTE', 'ESTU_LIMITA_MOTRIZ']

# Filtrar el DataFrame para mostrar solo las filas donde al menos una de las columnas de discapacidad no es nula
df_no_nulos_discapacidades = df[df[columnas_discapacidades].notna().any(axis=1)]

# Mostrar las primeras filas del DataFrame resultante
print(df_no_nulos_discapacidades.head())
```

```python
df['POSEE_INCAPACIDAD'] = df[columnas_discapacidades].notna().any(axis=1).astype(int)
df = df.drop(columnas_discapacidades, axis=1)
df.head()
```

```python
df.isna().sum()
```

```python
df['ESTU_TIENE_ETNIA'].value_counts()
```

```python
df['ESTU_ETNIA'].value_counts()
```

```python

# Crear la columna 'POSEE_ETNIA' basada en 'ESTU_TIENE_ETNIA'
df['POSEE_ETNIA'] = df['ESTU_TIENE_ETNIA'].apply(lambda x: 1 if x == 'SI' else 0)

# Identificar inconsistencias
inconsistencias = df[(df['POSEE_ETNIA'] == 1) & (df['ESTU_ETNIA'] == '100 Ninguno')]

# Mostrar las filas con inconsistencias
inconsistencias.head()
```

```python

# Identificar y corregir inconsistencias: si 'ESTU_ETNIA' es "NINGUNA", ajustar 'POSEE_ETNIA' a 0
df.loc[df['ESTU_ETNIA'] == '100 Ninguno', 'POSEE_ETNIA'] = 0

# Mostrar las filas con inconsistencias
inconsistencias = df[(df['POSEE_ETNIA'] == 1) & (df['ESTU_ETNIA'] == '100 Ninguno')]

# Mostrar las primeras filas con inconsistencias corregidas
print(inconsistencias[['ESTU_TIENE_ETNIA', 'ESTU_ETNIA', 'POSEE_ETNIA']].head())
```

```python
df = df.drop(['ESTU_TIENE_ETNIA', 'ESTU_ETNIA'], axis=1)
df.head()
```

Veamos a ver como se comportan las notas nulas

```python
columnas_notas = 'MOD_COMUNI_ESCRITA_PUNT'

sns.histplot(df[columnas_notas], kde=True)

# Agregar lÃ­neas para la media y la mediana
plt.axvline(df[columnas_notas].mean(), color='red', label='Media')
plt.axvline(df[columnas_notas].median(), color='green', label='Mediana')

plt.legend()
plt.show()
```

Como no tiene valor atipicos represtativos podemos cambiar los valores nulos por la media

```python
columnas_notas = ['MOD_COMUNI_ESCRITA_PUNT', 'MOD_COMUNI_ESCRITA_PNAL', 'MOD_COMUNI_ESCRITA_PGREF']
df[columnas_notas] = df[columnas_notas].apply(lambda x: x.fillna(x.median()))
# df[columnas_notas].fillna(df[columnas_notas].mean(), inplace=True)
df.isna().sum()
```

Hacmos los mismo para INSE y NES

```python
df[['NSE', 'INSE']].head()
```

```python

# Filtra los valores nulos
valores_limpios = [x for x in df['INSE'] if not pd.isnull(x)]

# Crea un histograma de los valores limpios
sns.histplot(valores_limpios, kde=True)

# Calcula la media y la mediana de los valores limpios
media = np.mean(valores_limpios)
mediana = np.median(valores_limpios)

# Agrega lÃ­neas para la media y la mediana en el grÃ¡fico
plt.axvline(media, color='red', label='Media')
plt.axvline(mediana, color='green', label='Mediana')

# Muestra la leyenda
plt.legend()

# Muestra el grÃ¡fico
plt.show()
```

```python
df['INSE'].fillna(mediana, inplace=True)
df.isna().sum()
```

```python
df['NSE'].value_counts()
```

```python
bins = np.linspace(min(df["INSE"]), max(df["INSE"]), 5)
bins
```

```python
group_names = df['NSE'].value_counts().index.values
group_names.sort()
group_names
```

```python
mascara_nulos = df['NSE'].isnull()
df.loc[mascara_nulos, 'NSE'] = pd.cut(df.loc[mascara_nulos, 'INSE'], bins=bins, labels=False, include_lowest=True)
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```
